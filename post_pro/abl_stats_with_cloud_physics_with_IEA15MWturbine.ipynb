```python
### Grab raw data from .nc file
with h5py.File(data_file_with_turbine_yaw_270_nc) as f:
    # Time
    time0 = f['time'][3600:7200]
    
    # xz-plane
    xy_coords_raw = f['xyplane']['coordinates'][:]

print(np.shape(time0))
print(np.shape(xy_coords_raw))

### Reformat xz plane
xy_xcoords = np.sort(np.unique(np.round(xy_coords_raw[:,0],6)))
xy_ycoords = np.sort(np.unique(np.round(xy_coords_raw[:,1],6)))
xy_zcoords = np.sort(np.unique(np.round(xy_coords_raw[:,2],6)))

xy_coords = {'time':time0, 'x': xy_xcoords, 'y': xy_ycoords, 'z': xy_zcoords}
ds_xy_with_turbine_yaw_270_nc = xr.Dataset(xy_coords)

for var in ['velocityx', 'velocityy', 'velocityz']:
    with h5py.File(data_file_with_turbine_yaw_270_nc, 'r') as f:
        curr_arr = f['xyplane'][var][3600:7200].astype(np.float32)
    arr_reshaped = curr_arr.reshape((len(time0), len(xy_zcoords), len(xy_ycoords), len(xy_xcoords)))
    arr_transposed = np.transpose(arr_reshaped, axes=[0,3,2,1])
    ds_xy_with_turbine_yaw_270_nc[var] = (('time', 'x', 'y', 'z'), arr_transposed)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rotate_coords(x, y, yaw_deg):
    """
    Rotate (x, y) coordinates by yaw_deg (degrees, positive CCW from x-axis).
    """
    theta = np.radians(yaw_deg)
    x_rot = x * np.cos(theta) + y * np.sin(theta)
    y_rot = -x * np.sin(theta) + y * np.cos(theta)
    return x_rot, y_rot

def rotate_velocity(u, v, yaw_deg):
    """
    Rotate (u, v) velocity components by yaw_deg (degrees, positive CCW from x-axis).
    """
    theta = np.radians(yaw_deg)
    u_rot = u * np.cos(theta) + v * np.sin(theta)
    v_rot = -u * np.sin(theta) + v * np.cos(theta)
    return u_rot, v_rot

# Example: Create a synthetic velocity field in the global frame
# (You would typically load this from simulation or measurement data)
x, y = np.meshgrid(np.linspace(-200, 800, 101), np.linspace(-200, 200, 51))
u = 8.0 * np.ones_like(x)  # Uniform inflow in x-direction
v = np.zeros_like(x)       # No cross-flow

# Yaw angle of the turbine (degrees, positive CCW from x-axis)
yaw_angle = 30

# Rotate coordinates and velocities to the turbine's local frame
x_rot, y_rot = rotate_coords(x, y, yaw_angle)
u_rot, v_rot = rotate_velocity(u, v, yaw_angle)

# Visualization: Plot the original and rotated velocity fields
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].quiver(x, y, u, v)
axs[0].set_title('Global Frame')
axs[0].set_xlabel('x [m]')
axs[0].set_ylabel('y [m]')
axs[0].axis('equal')

axs[1].quiver(x_rot, y_rot, u_rot, v_rot)
axs[1].set_title('Turbine-Aligned (Yawed) Frame')
axs[1].set_xlabel("x' [m]")
axs[1].set_ylabel("y' [m]")
axs[1].axis('equal')

plt.tight_layout()
plt.show()
```
